```python
#estados com mais municipios com nota máxima
import pandas as pd
df = pd.read_csv('EBT_20200601.csv', sep=';', encoding='cp1252')

df_sub = df[df['nota'] == 10.00 ]
df_sub['nome_uf'].value_counts().plot.bar(title='Quantidade de Municípios por estado com nota máxima por estado')
```

```python
import pandas as pd
df = pd.read_csv('EBT_20200601.csv', sep=';', encoding='cp1252')
df['regulamentou_lai'].value_counts().plot.pie(title= 'Municípios que regulamentaram a Lei de Acesso a Informacao')
df['regulamentou_lai'].value_counts()
```

```python
import pandas as pd
df = pd.read_csv('EBT_20200601.csv', sep=';', encoding='cp1252')
df_sub = df[(df['nota'] >= 5.00) & (df['nota'] <= 7.00)]
df_sub['nome_uf'].value_counts().plot.bar(title='Quantidade de Municípios por estado com nota maior ou igual a 5 e menor igual a 7 ')
df_sub['nome_uf'].value_counts()
```
